```python
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
import urllib.request
import plotly.io as pio
```

# Step 2: Reading the page

```python
url = 'https://en.wikipedia.org/wiki/Natural_language_processing'
```

```python
import ssl
context = ssl._create_unverified_context()
page = urllib.request.urlopen(url=url,context=context)
html_plain = page.read()
print(html_plain)
```

# Step 3: Data Cleaning

```python
soup = BeautifulSoup(html_plain,'html.parser')
soup_text = soup.get_text(strip = True)

ready_text = soup_text.lower()
print(ready_text)
```

# Step 4: Tokenization
splitting each word into a list item

```python
tokens = []
for t in ready_text.split():
    tokens.append(t)
print(tokens)
```

```python
#Run this line if you get an error message in the next code block
# nltk.download()

stop_words = stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stop_words:
        clean_tokens.remove(token)
print(clean_tokens)
```

# Step 5: Data Visualization
firstly we will count the frequency of the tokens and then we will filter the high-frequency ones

```python
#  calculate the frequency of words
freq = nltk.FreqDist(clean_tokens)
for key, val in freq.items():
    print('Word: ' + str(key) + ', Quantity:' + str(val))
```

```python
# get the tokens that has been used more than 10 times in the page
high_freq = dict()
for key, val in freq.items():
    if (val > 10):
        high_freq[key] = val

# !pip install nbformat 
```

```python
#Note: to pass keys and values of high_freq dictionary, I had to convert them to list when passing them
fig = dict({
    "data": [{"type": "bar",
              "x": list(high_freq.keys()),
              "y": list(high_freq.values())}],
    "layout": {"title": {"text": "Most frequently used words in the page"}, "xaxis": {"categoryorder":"total descending"}}
})
pio.show(fig)
```
